```python
 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("exams.csv")
data.head
```

```python
cdata = data[['math_score', 'reading_score']]
cdata.head()
```

```python
cdata.describe()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(cdata.math_score, cdata.reading_score)
plt.xlabel('Бал по математике')
plt.ylabel('Бал по чтению')
```

```python
model = LinearRegression()
x = pd.DataFrame(cdata.math_score)
y = pd.DataFrame(cdata.reading_score)
model.fit(x , y)
```

```python
from turtle import color

plt.figure(figsize=(10,6))
plt.scatter(cdata.math_score, cdata.reading_score)

plt.plot(x, model.predict(x), color = 'red')
plt.xlabel('Бал по математике')
plt.ylabel('Бал по чтению')
```

```python
model.coef_
```

```python
model.score(x, y)
```

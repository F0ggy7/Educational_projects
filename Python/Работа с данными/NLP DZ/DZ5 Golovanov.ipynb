```python
import string
import pymorphy2
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist

nltk.download('stopwords')
```

### 1. Найти вариант текста из Заданий 1-4 на русском языке

```python
# Читаем книгу и записываем ее в переменную.
with open('The Little Prince Ru.txt') as book:
    read_book = book.read()
```

### 2. Выполнить токенизацию

```python
# Токинизируем текс.
word_list = word_tokenize(read_book)
len(word_list)
```

### 3. Выполнить лемматизацию текста с использованием pymorphy, определить лексическое многообразие текста, самое часто используемое слово.

```python
# Удаляю знаки препинания и перевожу все слова в нижний регистр.
remove_punctuation = str.maketrans('', '', string.punctuation)
lower_text = [x for x in [t.translate(remove_punctuation).lower() for t in word_list] if len(x) > 0]
```

#### Выполнить лемматизацию текста с использованием pymorphy

```python
morph = pymorphy2.MorphAnalyzer()

normal_words = []
for word in lower_text:
    normal_word = morph.parse(word)[0].normal_form
    normal_words.append(normal_word)
normal_words
```

#### Определить лексическое многообразие текста

```python
lexical_diversity = (len(set(normal_words)) / len(normal_words)) * 100
print("Лексическое разнообразие текста составляет", lexical_diversity, "%")
```

#### Cамое часто используемое слово.

```python
# FreqDist Упорядочит слова по убыванию частоты использования
fdist = FreqDist(normal_words) 

data_word = []

# Заношу слова в список для обращения к ним.
for word in fdist:
    data_word.append(word)
most_frequent = data_word[0] 

print ("Самое частое по частоте использования слово это " "(", most_frequent,")")
```

```python
all_stopwords = stopwords.words('russian')
tokens_without_stop_w = [word for word in lower_text if not word in all_stopwords]
len(tokens_without_stop_w)
```

### *5. Написать свой pos-tagger для русского языка (минимум 2-3 части речи)

```python
vrb_end = ['ют', 'ут', 'у', 'ешь', 'ете', 'ем','ит', 'ят', 'ат', 'ет', 'ишь', 'ите',  'ю', 'им','у', 'ю']
vrb_inf_end = ['ть', 'ться', 'тся']
adj_end = ['ых','ий', 'ый', 'ого','ой', 'ее', 'ая', 'яя', 'ей', 'ую', 'юю', 'ие', 'ые', 'их', 'его', 'ому','им', 'ым', 'ом', 'ем', 'ое', 'ему', 'ими', 'ыми']
adv_beg = ['на', 'до', 'с' 'в','за', 'из', 'ис',]
adv_end = ['о', 'а']
```

```python
def ending(word, endings):
    for end in endings:
        if word.endswith(end):
            return True
    return False
    
def beging(word, endings):
    for end in endings:
        if word.startswith(end):
             return True
    return False
```

```python
def tager(data_word):
    res_data = []

    for word in data_word:
        res = []
        res.append(word)

        if ending(word, adv_end) and beging(word, adv_beg):
            res.append('ADV')

        elif ending(word, adj_end):
            res.append('ADJ')
            
        elif ending(word, vrb_end) or beging(word, vrb_inf_end):
            res.append('VERB')
                
        elif ('ADV' not in res) and not beging(word, vrb_inf_end):
            res.append('NOUN')

        res_data.append(res)
    return res_data
```

```python
print(tager(tokens_without_stop_w[100:115]))
```

```python
import nltk
from nltk import tokenize
from nltk.tokenize import *
from nltk.corpus import words
```

```python
# Читаем книгу и записываем ее в переменную
with open('The_Little_Prince.txt') as book:
    read_book = book.read()
```

```python
# Убераю из текста обозначение пустой строки
spec_chars = '\n' 
book = "".join([ch for ch in read_book if ch not in spec_chars])
```

### 1. Выбранный в Задании 1 текст разобрать по предложениям. Выбрать предложения с 100 по 115 включительно.

```python
tokens = sent_tokenize(book, language='english')
# Выделю предложения с 100 по 115
proposals = tokens[100: 115] 
# Занесу все в одну строку
text = ' '.join(proposals)
proposals
```

#### RegexpTokenizer

```python
# Этот токенизатор разбивает исходный текст на подстроки, используя регулярное выражение, передаваемое ему в качестве параметра.

# Патерн из буквенных последовательностей, денежных выражений и любых других последовательностей, отличных от пробелов:
pattern = '\w+|\$[\d\.]+|\S+'
print(RegexpTokenizer(pattern).tokenize(text))
```

```python
# gaps – логический параметр, определяющий, будет ли токенизатор искать сами токены (False) или разделители между ними (True).
print(RegexpTokenizer('\s+', gaps=True).tokenize(text))
```

##### Библиотека NLTK содержит также несколько подклассов RegexpTokenizer, использующих заранее определенные регулярные выражения: WhitespaceTokenizer, BlanklineTokenizer, WordPunctTokenizer.

```python
# исходный текст разделяется на группы символов, ограниченных пробелами
print(WhitespaceTokenizer().tokenize(text))
```

```python
# разбивает текст, рассматривая в качестве разделителя любую последовательность пустых строк
print(BlanklineTokenizer().tokenize(text))
```

```python
# Преобразует текст в последовательность буквенных и неалфавитных символов с помощью регулярного выражения ‘\w+|[^\w\s]+’
# Отличается от WhitespaceTokenizer тем, что отделяет знаки препинания
print(WordPunctTokenizer().tokenize(text))
```

#### SExprTokenizer


```python
# Этот токенизатор используется для поиска в исходном тексте выражений в скобках.
print(SExprTokenizer(strict=False).tokenize(text))
# По скольку в выборке нет скобок вывело всю выборку
```

#### Типа свой токинизатор

```python
def token(data_book):
    sentences = filter(lambda x: x, [t.strip() for t in re.compile(r'[.|!|?|…]').split(data_book)])
    data = []
    for s in sentences:
        data.append(s)
    return(data)
```

```python
token(book)
```

NLTK токинизатор выполняет токинизацию за 0.2с. Самопальный токинизатор за 0.1с. При этом самописный токинизатор справился с конкретно этой задачей лучше. Т.к в тексе присутсвовали аномалии после удаления \n. Так встречаются предложения которые идут сразу после точки пред идущего предложения Пример:(Привет.Привет). NLTK токинизатор прочитает это как одно предложение. Самописный же разделит на два разных предложения.

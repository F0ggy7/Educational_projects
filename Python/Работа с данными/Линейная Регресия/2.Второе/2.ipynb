```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl

from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("CAR DETAILS FROM CAR DEKHO.csv")
data.head
```

```python
cdata =data[['selling_price', 'km_driven']]
cdata.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(cdata.selling_price, cdata.km_driven, alpha=0.4)
plt.xlabel('Цена авто')
plt.ylabel('Пройденые километры')
```

```python
model = LinearRegression()
x = pd.DataFrame(cdata.selling_price)
y = pd.DataFrame(cdata.km_driven)
model.fit(x , y) 
```

```python
from turtle import color

plt.figure(figsize=(10,6))
plt.scatter(cdata.selling_price, cdata.km_driven, alpha=0.4)

plt.plot(x, model.predict(x), color = 'red')
plt.xlabel('Цена авто')
plt.ylabel('Пройденые километры')
```

```python
model.coef_
```

```python
model.score(x, y)
```

```python
import string
import nltk

from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer 
from nltk.probability import FreqDist

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Читаем книгу и записываем ее в переменную.
with open('The_Little_Prince.txt') as book:
    read_book = book.read()
```

```python
# Токинизируем текс.
word_list = word_tokenize(read_book)
len(word_list)
```

```python
# Удаляю знаки препинания и перевожу все слова в нижний регистр.
remove_punctuation = str.maketrans('', '', string.punctuation)
lower_text = [x for x in [t.translate(remove_punctuation).lower() for t in word_list] if len(x) > 0]
```

```python
# В тексте часто встречается символ � и машиной он читается как 'пїѕ'. Я рещил его добавить в список стоп слов.
all_stopwords = stopwords.words('english')
all_stopwords.append('пїѕ')
```

```python
# Удаляю стоп слова.
tokens_without_stop_w = [word for word in lower_text if not word in all_stopwords]
```

### 1. Выполнить лемматизацию текста из Задания 1 без указания части речи

```python
#Выполняю лемматизацию слов без пос тегирования. Использую Лемматизатор из nltk WorldNet
lemmatized_output = []

#Прохожусь по каждому слову списка и лемматизирую его.
for word in tokens_without_stop_w:
    lemmatized_word = WordNetLemmatizer().lemmatize(word)
    #Добавляю лемматизированное слово в список
    lemmatized_output.append(lemmatized_word)

lemmatized_output
```

### 2. Выполнить лемматизацию текста из Задания 1 с указанием части речи

```python
#Выполняю пос тегирование списка слов полученного из текста
pos_tag_word = pos_tag(tokens_without_stop_w)
pos_tag_word
```

```python
# get_pos был обучен с корпусом Treebank, получается он использует набор тегов Treebank. Но WorldNet использует собсвенные пос теги. 
# Меня немного удивило, что в nltk находятся обе эти функции, но связь между ними не проброшенно.
# Если в теге находится определеная буква, то он относится к определенной части речи. 
# Следующая функция сопоставит теги Treebank с именами частей речи WordNet.
def wordnet_pos(treebank_tag):
        if treebank_tag.startswith('R'):
            return wordnet.ADV
        elif treebank_tag.startswith('N'):
            return wordnet.NOUN
        elif treebank_tag.startswith('J'):
            return wordnet.ADJ
        elif treebank_tag.startswith('V'):
            return wordnet.VERB
        else:
            return wordnet.NOUN
```

```python
#Выполняю лемматизацию слов с учетом састи речи. 
lemmatized_output2 = []

for word in pos_tag_word:
    lemmatized_word = []
    pos = wordnet_pos(word[1])
    # Решил добавлять не только лемматизированное слово, но и к какой части речи оно относиться.
    lemmatized_word.append(WordNetLemmatizer().lemmatize(word[0], pos))
    lemmatized_word.append(pos)
    lemmatized_output2.append(lemmatized_word)

lemmatized_output2
```

### 4*. Написать свой лемматайзер для 20 самых частых по употреблению слов в тексте.

```python
# Опотребляю 20 самых частых по употреблению слов в тексте.
fdist = FreqDist(tokens_without_stop_w) 
data20 = []
# Заношу слова в список для обращения к ним.
fdist = fdist.most_common(20)
for word in fdist:
    data20.append(word[0])
    
print(data20)
```

Поскольку задание сделать лемматизатор для 20 самых частых слов в тексте будем хардкодить.

```python
def lemmatize(token):
  if len(token) < 2: # Если длина слова меньше двух, то это скорее всего что-то, что не нужно лемматизировать
    return token
  
  elif token.endswith('s'): # Проверяем окончание 
    if len(token) > 3:
      if token.endswith('s'):
        if len(token[:-3]) > 2:
          return token.replace('s','') #Заменяем окончание
  
  elif token == 'said':
      token = 'say'
      return token
  
  elif token == 'would':
      token = 'will'
      return token

  elif token == 'made':
      token = 'make'
      return token

  else:
    return token

  return token
```

```python
# Проверяем
lem_text = []
for word in data20:
    lem_word = lemmatize(word)
    lem_text.append(lem_word)

# Мой резульат :)
print(lem_text)
```
